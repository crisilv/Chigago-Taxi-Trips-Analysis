### Chicago Trip Analysis

With this project I will analyze Chiacago Taxi trip traslating SQL queries in Phyton pandas queries

```python
import pandas as pd
pd.set_option('display.max_rows', 100)
```

```python
df=pd.read_csv("Taxi_Final.csv", low_memory=False, thousands=',')
df["Trip Start Timestamp"] = pd.to_datetime(df["Trip Start Timestamp"], format="%m/%d/%Y %I:%M:%S %p")
df["Trip End Timestamp"] = pd.to_datetime(df["Trip End Timestamp"], format="%m/%d/%Y %I:%M:%S %p")
df.info()
```

```python
df.head()
```

```python
#Let's extract the data we need to analyze Datetime columns
df["Month"]=df['Trip Start Timestamp'].dt.month
df["Month_name"]=df['Trip Start Timestamp'].dt.month_name()
df["Day"]=df["Trip Start Timestamp"].dt.day
df['Date']=df['Trip Start Timestamp'].dt.date
df['Date'] = pd.to_datetime(df['Date'])
df[["Trip Start Timestamp","Month","Month_name","Day",'Date']]
```

```python
#Total trip for each month name
df.groupby(["Month","Month_name"], as_index=False)["Trip ID"].count().sort_values(by="Month", ascending=True)
```

```python
#Most common payment type
df.groupby("Payment Type", as_index=False)["Trip ID"].count().sort_values(by="Trip ID", ascending=False).rename(columns={"Trip ID" : "Total transaction"})
```

```python
#Most common payment type in january
#df.groupby("Payment Type", as_index=False)["Trip ID"].count().sort_values(by="Trip ID", ascending=False)
#df[df["Month"]==1].groupby("Payment Type", as_index=False)["Trip ID"].count().sort_values(by="Trip ID", ascending=False)
#df[df["Month"] == 1].groupby("Payment Type", as_index=False).agg(TripsNumber=('Trip ID', 'count'))
#df[df["Month"]==1].groupby("Payment Type", as_index=False).size().rename(columns={"size" : "Total transaction"})
df[df["Month"] == 1].groupby("Payment Type", as_index=False).agg(TripsNumber=('Trip ID', 'count')).sort_values(by="TripsNumber", ascending=False)
```

```python
#Average trip miles and seconds
a=df[["Trip Miles"]].mean().round(2).iloc[0]
b=((df[["Trip Seconds"]].mean()/60).round(2)).iloc[0]
print(f'Trip Miles: {a} \nTrip Seconds: {b} ')


```

```python
#Taxi demand prediction: calculation of the rolling average number of trips for each day over a period between the preceding 2 days and the following 2 days to see.
daily_trips=df.groupby("Date", as_index=False).agg(NumberOfTrips=('Trip ID', 'count'))
daily_trips["AvgNumberDaily"]=daily_trips["NumberOfTrips"].rolling(window=5, center=True).mean()
daily_trips.head()
```

```python
#For each community area return the number of each trip around the year
df["Rank"]=df.groupby("Pickup Community Area")['Trip Start Timestamp'].rank(method='first', ascending=True)
df[["Rank", "Pickup Community Area", 'Trip Start Timestamp' ]].sort_values(by=['Trip Start Timestamp',"Pickup Community Area"]).head(25)
```

```python
#Time elapsed between two different trips for the same taxi driver in minutes
df = df.sort_values(by=['Taxi ID', 'Trip Start Timestamp','Trip End Timestamp'])
df["PreviousEndTimestamp"]=df[df['Trip Miles']>0].groupby(["Taxi ID"], as_index=False)["Trip End Timestamp"].shift(1)
df["Time Elapsed"]=df["Trip Start Timestamp"]-df["PreviousEndTimestamp"]
df[df['Trip Miles']>0][["Taxi ID", "Trip Start Timestamp", "Trip End Timestamp", "PreviousEndTimestamp","Time Elapsed"]].sort_values(by=["Taxi ID", "Trip Start Timestamp",'Trip End Timestamp']).head(10)
```

```python
#Cumulative Trip Miles for each taxi
df = df.sort_values(by=['Taxi ID', 'Trip Start Timestamp','Trip End Timestamp'])
df["cumMiles"]=df.groupby(["Taxi ID"], as_index=False)["Trip Miles"].cumsum()
df[df["Trip Miles"]>0][["Taxi ID", "Date","Trip Start Timestamp", "Trip End Timestamp","Trip Miles","cumMiles"]].head(10)
```

```python
#Cumulative Trip Miles for each taxi for each day
df = df.sort_values(by=['Taxi ID', 'Trip Start Timestamp','Trip End Timestamp'])
df["cumMiles"]=df.groupby(["Taxi ID", "Month","Month_name"], as_index=False)["Trip Miles"].cumsum()
df[df["Trip Miles"]>0][["Taxi ID", "Month","Month_name","Trip Start Timestamp", "Trip End Timestamp","Trip Miles","cumMiles"]].head(10)
```

```python
#Cumulative number of trips for each taxi, each day
df=df.sort_values(by=["Taxi ID",'Trip Start Timestamp','Trip End Timestamp' ])
df["CumNumb"]=df[df["Trip Miles"]>0].groupby(["Taxi ID", 'Date'], as_index=False)["Trip ID"].cumcount()+1
df[df["Trip Miles"]>0][["Taxi ID", "Date","Trip Start Timestamp", "Trip End Timestamp","Trip Miles","CumNumb"]].head(10)
```

```python
#First Pickup and Last Dropoff community Area for each taxi each day
df=df.sort_values(by=["Taxi ID",'Trip Start Timestamp','Trip End Timestamp'])
#df[df["Trip Miles"]>0].groupby(["Taxi ID", "Date"]).agg(firstPickup=('Pickup Community Area','first'), lastPick=('Dropoff Community Area','last')  ).head(10)
df["FirstPickup"]=df[df['Trip Miles']>0].groupby(["Taxi ID", "Date"])['Pickup Community Area'].transform("first")
df["LastPickup"]=df[df['Trip Miles']>0].groupby(["Taxi ID", "Date"])['Dropoff Community Area'].transform("last")
df[df['Trip Miles']>0][["Taxi ID", "Date","Trip Start Timestamp", "Trip End Timestamp","FirstPickup",'Pickup Community Area',"LastPickup",'Dropoff Community Area']].head(25)
# If I wanna see the daily report
# df[df['Trip Miles']>0][["Taxi ID", "Date","FirstPickup","LastPickup"]].drop_duplicates()
```

```python
#Taxi with the most number of miles done
MilesXTaxi=df.groupby("Taxi ID", as_index=False)["Trip Miles"].sum()
MilesXTaxi[MilesXTaxi["Trip Miles"]==MilesXTaxi["Trip Miles"].max()]
#MilesXTaxi.sort_values(by=["Trip Miles"], ascending=False).head(1)
```

```python
#Taxi with the most number of trips done for each month
MonthlyTaxi=df.groupby(['Taxi ID', 'Month', 'Month_name'], as_index=False).agg(NumberOfTrips=('Taxi ID', 'count'))
Jan=MonthlyTaxi[MonthlyTaxi["Month"]==1]['NumberOfTrips'].max()
Feb=MonthlyTaxi[MonthlyTaxi["Month"]==2]['NumberOfTrips'].max()
Mar=MonthlyTaxi[MonthlyTaxi["Month"]==3]['NumberOfTrips'].max()
MonthlyTaxi[((MonthlyTaxi["Month"]==1) & (MonthlyTaxi["NumberOfTrips"]==Jan)) | ((MonthlyTaxi["Month"]==2) & (MonthlyTaxi["NumberOfTrips"]==Feb)) | ((MonthlyTaxi["Month"]==3) & (MonthlyTaxi["NumberOfTrips"]==Mar)) ]
```

```python
#Report with: company with max number of trips, company with min number of trips, company with max number of miles, company with min number of miles
report=df[df['Date']<'2021-01-12'].groupby("Company", as_index=False).agg(TotalTrips=('Trip ID','count'), TotalMiles=('Trip Miles',lambda x: round(x.sum(),0)))
MaxTrips=report["TotalTrips"].max()
LeastTrips=report["TotalTrips"].min()
MaxMiles=report["TotalMiles"].max()
MinMiles=report[report["TotalMiles"]>0]["TotalMiles"].min()
CompMaxTrip=report[report["TotalTrips"]==MaxTrips].groupby("TotalTrips", as_index=False)["Company"].agg('/'.join).loc[0,"Company"]
CompMinTrip=report[report["TotalTrips"]==LeastTrips].groupby("TotalTrips", as_index=False)["Company"].agg('/'.join).loc[0,"Company"]
CompMaxMiles=report[report["TotalMiles"]==MaxMiles].groupby("TotalMiles", as_index=False)["Company"].agg('/'.join).loc[0,"Company"]
CompMinMiles=report[report["TotalMiles"]==MinMiles].groupby("TotalMiles", as_index=False)["Company"].agg('/'.join).loc[0,"Company"]
companies=[CompMaxTrip,CompMinTrip,CompMaxMiles,CompMinMiles]
metric=['Companies with most trips done', 'Companies with least trips done', 'Companies with most miles done', 'Companies with least miles done']
values=[str(MaxTrips)+' Trips', str(LeastTrips)+' Trips', str(MaxMiles)+' Miles', str(MinMiles)+' Miles']
result=pd.DataFrame({'Metric': metric, 'Companies':companies, 'Value':values})
result
```
